```python
## The second function that I will write will be the function to check if the game_board is full. If it is, than it's a tie and nobody won the game. 

def board_full(gaming_board):
    '''Input: board game
    Output: True or False depending on if the board game is full or not '''
    for key in gaming_board.keys():
        if gaming_board[key] == ' ':
            return False
    else:
        print("Game over. It's a tie")
        return True
```
