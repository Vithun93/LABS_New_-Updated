```python
import requests
```

```python
url="https://www.reuters.com/world/asia-pacific/pelosi-expected-arrive-taiwan-tuesday-sources-say-2022-08-02/"
html=requests.get(url).content
html[0:600]
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(html,"lxml")
soup
```

```python
tags = ["h1","h2", "h3", "h4","h5","h6","h7","p"]
text=[element.text for element in soup.find_all(tags)]
text
```

```python
print("\n ".join(text))
```
