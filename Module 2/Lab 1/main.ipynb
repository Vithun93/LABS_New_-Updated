```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
fig, [ax1, ax2] = plt.subplots(2, 1, figsize=(12,4))
med_monthly_steps = fitbit.groupby("Months_encoded")["Steps"].median()
x = pd.DataFrame(med_monthly_steps).reset_index()
ax1.plot(x["Months_encoded"], x["Steps"])
ax1.set_xlabel("Month")
ax1.set_ylabel("Median steps")
ax1.set_title("Median steps per month", fontsize=17)
# MEDIAN STEPS BY WORK_OR_WEEKEND
med_work_or_weekend = fitbit.groupby("Work_or_Weekend")["Steps"].median()
y = pd.DataFrame(med_work_or_weekend).reset_index()
ax2.plot(y["Work_or_Weekend"], y["Steps"])
ax2.set_xlabel("Work/Weekend")
ax2.set_ylabel("Median steps")
ax2.set_title("Median steps per work or weekday", fontsize=17)
```
