```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
fig, [ax1, ax2] = plt.subplots(2, 1, figsize=(12,4))
med_monthly_steps = fitbit.groupby("Months_encoded")["Steps"].median()
x = pd.DataFrame(med_monthly_steps).reset_index()
ax1.plot(x["Months_encoded"], x["Steps"])
ax1.set_xlabel("Month")
ax1.set_ylabel("Median steps")
ax1.set_title("Median steps per month", fontsize=17)
# MEDIAN STEPS BY WORK_OR_WEEKEND
med_work_or_weekend = fitbit.groupby("Work_or_Weekend")["Steps"].median()
y = pd.DataFrame(med_work_or_weekend).reset_index()
ax2.plot(y["Work_or_Weekend"], y["Steps"])
ax2.set_xlabel("Work/Weekend")
ax2.set_ylabel("Median steps")
ax2.set_title("Median steps per work or weekday", fontsize=17)
```

#### Now in addition to the summary statistics, we want to have in the same plot the distribution of `Age`. What kind of plot would you use? Plot it. 

```python
# your code here
# your code here
fig = plt.figure()
sns.displot(titanic, x='Age')
mean=titanic['Age'].mean()
median=titanic['Age'].median()
std=titanic['Age'].std()
plt.vlines(mean, 0, mean, color='crimson',linewidth=4, label = 'mean')
plt.vlines(median, 0, median, color='green',linewidth=4, label = 'median')
plt.vlines(std, 0, std, color='c',linewidth=4, label = 'standard deviation')
plt.legend(loc=1, prop={'size':10})
plt.title('Age distribution with basic statistics')
plt.show()
```

#### What additional information does the last plot provide about feature `Age`?

```python
"""
It allows to compare age distribution with the basic statistics. For example the median and the mean are very close to the mode value (the highest blue bar); They also appear at the center of the distribution.
"""
```
